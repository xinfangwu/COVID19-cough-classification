```python
import os, random, shutil
```

```python
def movefile(filedir, tardir):
    pathdir = os.listdir(filedir)
    filenumber = len(pathdir)
    rate = 0.2
    picknumber = int(filenumber * rate)
    sample = random.sample(pathdir, picknumber)
    print(sample)
    for name in sample:
        shutil.move(filedir+  name, tardir+name)
    return
```

```python
filedir = r'M:\cs230_project\virufy_small_dataset_12s_classification\train\nondetectable\\'

tardir = r'M:\cs230_project\virufy_small_dataset_12s_classification\test\nondetectable\\'
movefile(filedir, tardir)
```

```python
filedir = r'M:\cs230_project\virufy_small_dataset_12s_classification\train\detectable\\'

tardir = r'M:\cs230_project\virufy_small_dataset_12s_classification\test\detectable\\'
movefile(filedir, tardir)
```
